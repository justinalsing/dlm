```python
# Import stuff
import pystan
import matplotlib.pyplot as plt
import numpy as np
import pickle
import time
import scipy.optimize as optimize
import scipy.interpolate as interpolate
import netCDF4
import mlr
%matplotlib inline
```

```python
code = """

data {
    int N; // number of time steps
    int nreg; // number of regressors
    vector[N] d; // n-composites vectors of N-time-steps -- the data
    vector[N] s; // corresponding standard deviation for each data point
    vector[nreg] regressors[N]; // nreg regressors of N-time-steps
    matrix[4,4] G; // Seasonal transition matrix
    real<lower=0> sigma_trend_prior;
    real<lower=0> sigma_seas_prior;
    real<lower=0> sigma_AR_prior;
}

parameters {
    vector[nreg] b; // regression coefficients
    vector[N] ar; // ar-process
    vector[4] seasonal_components[N]; // seasonal cycle
    vector[N] slope; // dynamical linear trend [slope]
    real x0;
    real<lower=0> sigma_trend;
    real<lower=0> sigma_seas;
    real<lower=0> sigma_AR;
    real<lower=0, upper=1> rho;
}

transformed parameters{
    vector[N] trend; // dynamical linear trend
    vector[N] seasonal; // seasonal cycle
    vector[N] regression_model; // complete regression model

    // Model for the latent dynamical model
    trend[1] = x0;
    seasonal[1] = seasonal_components[1][1] + seasonal_components[1][3];
    regression_model[1] = dot_product(b, regressors[1]);
    for(t in 2:N){

        trend[t] = trend[t-1] + slope[t];
        seasonal[t] = seasonal_components[t][1] + seasonal_components[t][3];
        regression_model[t] = dot_product(b, regressors[t]);
    }
}

model {
 
    // Priors on hyper parameters
    sigma_trend ~ normal(0, sigma_trend_prior);
    sigma_seas ~ normal(0, sigma_seas_prior);
    sigma_AR ~ normal(0, sigma_AR_prior);
    seasonal_components[1] ~ normal(0, 10);
    rho ~ uniform(0, 1);
    
    // Prior on AR process initial value
    ar[1] ~ normal(0, sqrt(sigma_AR^2/(1-rho^2)));

    // Model for the latent dynamical model
    for(t in 2:N){
    
        // AR-process
        ar[t] ~ normal(ar[t-1]*rho, sigma_AR);
        
        // Seasonal cycle
        seasonal_components[t] ~ normal(G*seasonal_components[t-1], sigma_seas);

        // Stochastic DLM trend
        slope[t] ~ normal(slope[t-1], sigma_trend);
    }

    # Construct target density
    for(t in 1:N){
        target += normal_lpdf(ar[t] + trend[t] + seasonal[t] + regression_model[t] | d[t], s[t]);
    }
}

"""
```

```python
# Compile the model
model = pystan.StanModel(model_code=code)
```

```python
# Import data
data = netCDF4.Dataset('data/BASIC_V1_2016_lotus_seascyc.nc')

# Which pressure and latitude index do you want to analyze?
pressure = 12
latitude = 8

# Time, pressure and latitude variables
T = data['time'][:]
P = data['pressure'][:]
L = data['latitude'][:]

# How many time steps are there?
N = len(T)

# Pick out a pressure and latitude panel: this is the "data" time-series
d = data['o3'][:, pressure, latitude]

# Extract the error on the time-series
s = data['o3_sigma'][:, pressure, latitude]

# Re-scale the data so it is -0.5 to 0.5, apply same scaling transform to error bars
scale = max(d) - min(d)
d = d - np.mean(d)
d = d/scale
s = s/scale

# Plot the data and error bars
plt.title('Pressure {} hPa; Latitude {} deg'.format(P[pressure], L[latitude]))
plt.plot(T, d, color = 'red')
plt.fill_between(T, d - s, d + s, color = 'grey', alpha = 0.5)
plt.xlabel('time, T [Julian date]')
plt.ylabel('Ozone')
plt.show()
```

```python
# Import and rescale all the regressors, extract the bit over the time period you actually want

# ENSO
y = np.loadtxt('regressors/ENSO_MEI_ENSO_1950_201704.dat')
t = np.loadtxt('regressors/ENSOTime_MEI_ENSO_1950_201704.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
enso = Y(T) # This extacts the regressor at the times you have in your dataset, interpolated in case the time grid is different

# Solar
y = np.loadtxt('regressors/Flux_F30_monthly_195111_201706_absolute.dat')
t = np.loadtxt('regressors/Time_F30_monthly_195111_201706_absolute.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
solar = Y(T)

# QBO30
y = np.loadtxt('regressors/multi_qbo30_1953_2016.dat')
t = np.loadtxt('regressors/multi_qbotime_1953_2016.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
qbo30 = Y(T) 

# QBO50
y = np.loadtxt('regressors/multi_qbo50_1953_2016.dat')
t = np.loadtxt('regressors/multi_qbotime_1953_2016.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
qbo50 = Y(T) # This extacts the regressor at the times you have in your dataset, interpolated in case the time grid is different

# SAOD
y = np.loadtxt('regressors/Reg_SAODak_1511979_15122015.dat')
t = np.loadtxt('regressors/Time_SAODak_1511979_15122015.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
saod = Y(T)

# Stack the regressors into big 2d array
nreg = 5 # number of regressors
regressors = np.column_stack([enso, solar, qbo30, qbo50, saod])

# Plot all the regressors
plt.plot(T, enso)
plt.show()

plt.plot(T, solar)
plt.show()

plt.plot(T, qbo30)
plt.show()

plt.plot(T, qbo50)
plt.show()

plt.plot(T, saod)
plt.show()
```

```python
# Declare matrix for seasonal cycle updates
G = np.zeros((4,4))
G[0:2, 0:2] = np.array([[np.cos(2*np.pi/12), np.sin(2*np.pi/12)],[-np.sin(2*np.pi/12), np.cos(2*np.pi/12)]])
G[2:4, 2:4] = np.array([[np.cos(2*2*np.pi/12), np.sin(2*2*np.pi/12)],[-np.sin(2*2*np.pi/12), np.cos(2*2*np.pi/12)]])
```

```python
# Do a cheeky MLR for initializing the DLM
ar_process, beta_coefficients, trend_slope, seasonal_cycle, sigma_AR, rho_AR = mlr.mlr_initialization(regressors.T, nreg, T, d)
```

```python
# Set the data and initialization of parameters for the DLM

# Define the data
data = {
                    'd':d,
                    's':s,
                    'N':N,
                    'nreg':nreg,
                    'regressors':regressors,
                    'G':G,
                    'sigma_trend_prior':0.01,
                    'sigma_seas_prior':0.01,
                    'sigma_AR_prior':0.5
                }

# Initialization
initial_state = {'ar':ar_process,
                 'b':beta_coefficients[0:nreg], 
                 'slope':trend_slope, 
                 'x0':0,
                 'seasonal_components':seasonal_cycle,
                 'sigma_trend':0.001,
                 'sigma_seas':0.001,
                 'sigma_AR':sigma_AR,
                 'rho':rho_AR
                }
```

```python
# Do the sampling
start = time.time()
fit = model.sampling(data=data, iter=2000, chains=1, init = [initial_state])
print(time.time() - start)
```

```python
# Plot recovered trend
plt.plot(T, np.mean(fit.extract()['trend'][:,:], axis = 0), color = 'red', ls = '--')
plt.fill_between(T, np.mean(fit.extract()['trend'][:,:], axis = 0) - np.std(fit.extract()['trend'][:,:], axis = 0), np.mean(fit.extract()['trend'][:,:], axis = 0) + np.std(fit.extract()['trend'][:,:], axis = 0), color = 'grey', alpha = 0.5)
plt.show()
```

```python
# Plot histograms of the beta coefficients
for i in range(nreg):
    plt.hist(fit.extract()['b'][:,i], bins=20, normed=True, alpha = 0.5)
    plt.show()
```

```python
# Plot seasonal cycle
plt.plot(T, np.mean(fit.extract()['seasonal'][:,:], axis = 0))
plt.show()
```

```python
# How did burn in go for, say, the last value of the trend? [it looks pretty well burnt in]
plt.plot(fit.extract()["trend"][:,-1])
plt.show()
```

```python
# Posteriors for the hyper parameters
plt.hist(fit.extract()['sigma_trend'], bins=20, normed=True, alpha = 0.5)
plt.show()

plt.hist(fit.extract()['sigma_seas'], bins=20, normed=True, alpha = 0.5)
plt.show()

plt.hist(fit.extract()['sigma_AR'], bins=20, normed=True, alpha = 0.5)
plt.show()

plt.hist(fit.extract()['rho'], bins=20, normed=True, alpha = 0.5)
plt.show()
```

```python
code_kalman = """

data {
    int N; // number of time steps
    int nreg; // number of regressors
    int nx; // length of state vector
    vector[N] d; // n-composites vectors of N-time-steps -- the data
    vector[N] s; // corresponding std deviation for each data point
    vector[nreg] regressors[N]; // nreg regressors of N-time-steps
    real<lower=0> sigma_trend_prior;
    real<lower=0> sigma_seas_prior;
    real<lower=0> sigma_AR_prior;
    real<lower=0> S;
}

transformed data {
    
    // Declare F-vector (observation projection vector)
    vector[nx] F[N+1];
    
    // Initialize
    for(t in 1:N+1){
        F[t] = rep_vector(0, nx);
    }
    
    // First time point
    F[1][1] = 1;
    F[1][2] = 0;
    F[1][3] = 1;
    F[1][4] = 0;
    F[1][5] = 1;
    F[1][6] = 0;
    for(i in 1:nreg){
        F[1][6+i] = 0; 
    }  
    F[1][6+nreg+1] = 1;
    
    // Loop over next time points
    for(t in 2:N+1){
        F[t][1] = 1;
        F[t][2] = 0;
        F[t][3] = 1;
        F[t][4] = 0;
        F[t][5] = 1;
        F[t][6] = 0;
        for(i in 1:nreg){
            F[t][6+i] = regressors[t-1][i]; 
        }  
        F[t][6+nreg+1] = 1;
    }
}

parameters {
    real<lower=0> sigma_trend;
    real<lower=0> sigma_seas;
    real<lower=0> sigma_AR;
    real<lower=0, upper=1> rhoAR;
}

transformed parameters{

    // Declare intermediate Kalman objects
    vector[nx] x_hat[N+1];
    vector[nx] x_bar[N+1];
    matrix[nx,nx] C_hat[N+1];
    vector[N+1] C_y;
    vector[nx] K[N+1];
    matrix[nx,nx] C_bar[N+1];
    vector[N+1] v;

    // Declare state-space model matrices
    matrix[nx, nx] G; // G-matrix (state transition matrix)
    matrix[nx, nx] W; // W-matrix (state covariance)
    matrix[nx, nx] C; // C-matrix (observation covariance)
    
    // Initialize state-space matrices
    G = rep_matrix(0, nx, nx);
    W = rep_matrix(0, nx, nx);
    C = rep_matrix(0, nx, nx);
    
    // Initialize Kalman filter stuff
    C_y = rep_vector(0, N+1);
    v = rep_vector(0, N+1);
    for(t in 1:N+1){
        x_hat[t] = rep_vector(0, nx);
        x_bar[t] = rep_vector(0, nx);
        K[t] = rep_vector(0, nx);
        C_bar[t] = rep_matrix(0, nx, nx);
        C_hat[t] = rep_matrix(0, nx, nx);
    }
    
    // Assign the G matrix
    
    // Trend part
    G[1,1] = 1;
    G[1,2] = 1;
    G[2,1] = 0;
    G[2,2] = 1;
    
    // 12-month seasonal part
    G[3,3] = cos(2*pi()/12);
    G[3,4] = sin(2*pi()/12);
    G[4,3] = -sin(2*pi()/12);
    G[4,4] = cos(2*pi()/12);
    
    // 6-month seasonal part
    G[5,5] = cos(2*pi()/6);
    G[5,6] = sin(2*pi()/6);
    G[6,5] = -sin(2*pi()/6);
    G[6,6] = cos(2*pi()/6);
    
    // Regressor part
    for(i in 1:nreg){
        G[6+i,6+i] = 1;
    }
    
    // AR process part
    G[6+nreg+1,6+nreg+1] = rhoAR;
    
    // Assign W matrix
    W[1,1] = 1e-10;
    W[2,2] = sigma_trend^2;
    W[3,3] = sigma_seas^2;
    W[4,4] = sigma_seas^2;
    W[5,5] = sigma_seas^2;
    W[6,6] = sigma_seas^2;
    for(i in 1:nreg){
        W[6+i,6+i] = 1e-10;
    }
    W[nx,nx] = sigma_AR^2;
    
    // Assign C matrix
    
    // Prior covariance of regression coefficients (and seasonal cycle and trend amplitudes)
    for(i in 1:(nx-1)){
        C[i,i] = S;
    }
    
    // Prior (stationary) covariance of the AR1 process
    C[nx,nx] = sigma_AR^2/(1-rhoAR^2);
    
    // Kalman filtering
    
    // Initialize C_bar
    C_bar[1] = C;
        
    // Do the forward filtering
    for(t in 2:N+1){
        x_hat[t] = G*x_bar[t-1];
        C_hat[t] = G*(C_bar[t-1]*G') + W;
        C_y[t] = F[t]'*C_hat[t]*F[t] + s[t-1]^2;
        K[t] = (C_hat[t]*F[t])/C_y[t];
        v[t] = d[t-1] - F[t]'*x_hat[t];
        x_bar[t] = x_hat[t] + K[t]*v[t];
        C_bar[t] = C_hat[t] - (K[t]*F[t]')*C_hat[t];
    }
}

model {
 
    // Priors on hyper parameters
    sigma_trend ~ normal(0, sigma_trend_prior);
    sigma_seas ~ normal(0, sigma_seas_prior);
    sigma_AR ~ normal(0, sigma_AR_prior);
    rhoAR ~ uniform(0, 1);
    
    // Construct target density
    for(t in 2:N+1){
        target += normal_lpdf(v[t] | 0, sqrt(C_y[t]));
    } 
}

generated quantities {

    // Declare Kalman smoother quantities
    vector[N] trend;
    vector[N] seasonal;
    vector[N] ar;
    vector[nx] x[N];
    vector[nx] x_uhat[N+1];
    vector[nx] x_ubar[N+1];
    vector[nx] x_uhat_hat[N+1];
    vector[N+1] y_uhat;
    vector[N+1] v_uhat;
    vector[nx] r[N+2];
    vector[nx] r_uhat[N+2];
    matrix[nx,nx] L[N+2];
    matrix[nx,nx] M[N+2];
    matrix[nx,nx] C_twidle[N+1];
    vector[nx] x_twidle[N+1];
    vector[nx] x_uhat_twidle[N+1];
    
    // Initialize matrix quantities and end points of Kalman smoothing vectors
    for(t in 1:N+1){
        L[t] = rep_matrix(0, nx, nx);
        M[t] = rep_matrix(0, nx, nx);
        C_twidle[t] = rep_matrix(0, nx, nx);
    }
    L[N+2] = rep_matrix(0, nx, nx);
    M[N+2] = rep_matrix(0, nx, nx);
    r[N+2] = rep_vector(0, nx);
    r_uhat[N+2] = rep_vector(0, nx);
 
    // Generate a draw from the state-space equations
    x_uhat[1] = multi_normal_rng(rep_vector(0, nx), C);
    y_uhat[1] = F[1]'*x_uhat[1] + normal_rng(0, 10*s[1]);
    for(t in 2:N+1){
        x_uhat[t] = G*x_uhat[t-1] + multi_normal_rng(rep_vector(0, nx), W);
        y_uhat[t] = F[t]'*x_uhat[t] + normal_rng(0, s[t-1]);
    }
    
    // Generate Kalman filter of realization; x_uhat_hat
    x_ubar[1] = rep_vector(0, nx);
    for(t in 2:N+1){
        x_uhat_hat[t] = G*x_ubar[t-1];
        v_uhat[t] = y_uhat[t] - F[t]'*x_uhat_hat[t];
        x_ubar[t] = x_uhat_hat[t] + K[t]*v_uhat[t];
    }
    
    // Generate Kalman smoother of realization and of data
    for(t in 1:N+1){
    
        L[N+2-t] = G - G*(K[N+2-t]*F[N+2-t]');
        r[N+2-t] = F[N+2-t]*v[N+2-t]/C_y[N+2-t] + L[N+2-t]'*r[N+2-t+1];
        r_uhat[N+2-t] = F[N+2-t]*v_uhat[N+2-t]/C_y[N+2-t] + L[N+2-t]'*r_uhat[N+2-t+1];
        M[N+2-t] = (F[N+2-t]*F[N+2-t]')/C_y[N+2-t] + L[N+2-t]'*(M[N+2-t+1]*L[N+2-t]);
        x_twidle[N+2-t] = x_hat[N+2-t] + C_hat[N+2-t]*r[N+2-t];
        x_uhat_twidle[N+2-t] = x_uhat_hat[N+2-t] + C_hat[N+2-t]*r_uhat[N+2-t];
        C_twidle[N+2-t] = C_hat[N+2-t] - C_hat[N+2-t]*(M[N+2-t]*C_hat[N+2-t]);        
    }
    
    for(t in 1:N){
        x[t] = x_uhat[t+1] - x_uhat_twidle[t+1] + x_twidle[t+1];
        trend[t] = x[t][1];
        seasonal[t] = x[t][3] + x[t][5];
        ar[t] = x[t][nx];
    }
}

"""
```

```python
# Compile the model
model_kalman = pystan.StanModel(model_code=code_kalman)

```

```python
# Set the data and initialization of parameters for the DLM

# Define the data
data_kalman = {
                    'd':d,
                    's':s,
                    'N':N,
                    'nx':nreg+4+2+1,
                    'nreg':nreg,
                    'regressors':regressors,
                    'S':10,
                    'sigma_trend_prior':0.01,
                    'sigma_seas_prior':0.01,
                    'sigma_AR_prior':0.5
                }

# Initialization
initial_state_kalman = {
                 'sigma_trend':0.0001,
                 'sigma_seas':0.001,
                 'sigma_AR':0.075,
                 'rhoAR':0.6
                }
```

```python
start = time.time()
fit = model_kalman.sampling(data=data_kalman, iter=2000, chains=1, init = [initial_state_kalman], verbose=True, pars=('sigma_trend', 'sigma_seas', 'sigma_AR', 'rhoAR', 'trend'))
print(time.time() - start)
```

```python
# Plot recovered trend
plt.plot(T, np.mean(fit.extract()['trend'][:,:], axis = 0), color = 'red', ls = '--')
plt.fill_between(T, np.mean(fit.extract()['trend'][:,:], axis = 0) - np.std(fit.extract()['trend'][:,:], axis = 0), np.mean(fit.extract()['trend'][:,:], axis = 0) + np.std(fit.extract()['trend'][:,:], axis = 0), color = 'grey', alpha = 0.5)
plt.show()
```
