```python
# Import stuff
import pystan
import matplotlib.pyplot as plt
import numpy as np
import pickle
import time
import scipy.optimize as optimize
import scipy.interpolate as interpolate
import netCDF4
from stan_dlm_models import *
import scipy.stats as stats
%matplotlib inline
```

```python
# Create seasonal GP
x = np.arange(1,13)
l = 1.4
A = 1
nreg = 5

# Covariance
C = np.zeros((12, 12))
for i in range(0, 12):
    for j in range(0,12):
        C[i,j] = A*np.exp(-(min(abs(i-j), abs(i-j-12), abs(i-j+12)))**2/(2*l**2))

C_GP_prior = np.kron(np.eye(nreg), C)

L = np.linalg.cholesky(C)
plt.imshow(C_GP_prior)
plt.show()

for i in range(0, 3):
    Y = interpolate.InterpolatedUnivariateSpline(x, np.dot(L, np.random.normal(0, 1, 12)), k = 3)
    plt.plot(np.linspace(1, 12, 100), Y(np.linspace(1, 12, 100)))
plt.show()
```

```python
# Compile the DLM model
model_seasonal_ar1 = pystan.StanModel(model_code=code_kalman_seasonal_ar1)
```

```python
# Import data
data = netCDF4.Dataset('../../bayesian_composite/iron/results/BASIC_V1_2017_lotus_seascyc_gcsw2017_fac2.nc')

# Which pressure and latitude index do you want to analyze?
pressure = -12
latitude = -1

# Time, pressure and latitude variables
T = data['time'][:]
P = data['pressure'][:]
L = data['latitude'][:]

# How many time steps are there?
N = len(T)

# Pick out a pressure and latitude panel: this is the "data" time-series
d = data['o3'][:, pressure, latitude]

# Extract the error on the time-series
s = data['o3_sigma'][:, pressure, latitude]

# Re-scale the data so it is -0.5 to 0.5, apply same scaling transform to error bars
scale = max(d) - min(d)
d = d - np.mean(d)
d = d/scale
s = s/scale

# Plot the data and error bars
plt.title('Pressure {} hPa; Latitude {} deg'.format(P[pressure], L[latitude]))
plt.plot(T, d, color = 'red')
plt.fill_between(T, d - s, d + s, color = 'grey', alpha = 0.5)
plt.xlabel('time, T [Julian date]')
plt.ylabel('Ozone')
plt.show()
```

```python
# Import and rescale all the regressors, extract the bit over the time period you actually want

# ENSO
y = np.loadtxt('regressors/ENSO_MEI_ENSO_1950_201802.dat')
t = np.loadtxt('regressors/ENSOTime_MEI_ENSO_1950_201802.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
enso = Y(T) # This extacts the regressor at the times you have in your dataset, interpolated in case the time grid is different

# Solar
y = np.loadtxt('regressors/Flux_F30_monthly_195111_201803_absolute.dat')
t = np.loadtxt('regressors/Time_F30_monthly_195111_201803_absolute.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
solar = Y(T)

# QBO30
y = np.loadtxt('regressors/multi_qbo30_1953_2018.dat')
t = np.loadtxt('regressors/multi_qbotime_1953_2018.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
qbo30 = Y(T) 

# QBO50
y = np.loadtxt('regressors/multi_qbo50_1953_2018.dat')
t = np.loadtxt('regressors/multi_qbotime_1953_2018.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
qbo50 = Y(T) # This extacts the regressor at the times you have in your dataset, interpolated in case the time grid is different

# AAO
y = np.loadtxt('regressors/Reg_aao_197901_201802.txt')
t = np.loadtxt('regressors/Time_aao_197901_201802.txt')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
aao = Y(T) # This extacts the regressor at the times you have in your dataset, interpolated in case the time grid is different

# SAOD
saod_data = netCDF4.Dataset('regressors/sad_1979_2017_10deg_60S_60N_missNans.nc')
y = saod_data['saod_strat_column'][:, latitude]
t = saod_data['time'][:]
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
saod = Y(T)

# Stack the regressors into big 2d array
nreg = 5 # number of regressors
regressors = np.column_stack([enso, solar, qbo30, qbo50, saod])

# Plot all the regressors
plt.plot(T, enso)
plt.show()

plt.plot(T, solar)
plt.show()

plt.plot(T, qbo30)
plt.show()

plt.plot(T, qbo50)
plt.show()

plt.plot(T, saod)
plt.show()

plt.plot(T, aao)
plt.show()
```

```python
# Construct seasonal regressors

# Compute the mask
mask = np.zeros((12, len(T)))
for i in range(0, 12):
    mask[i,np.arange(0,len(T)) % 12 == i] = 1

# Set-up the masked regressors
monthly_regressors = np.zeros((len(T), nreg*12))
for r in range(nreg):
    for i in range(0, 12):
        monthly_regressors[:, r*12 + i] = regressors[:,r]*mask[i,:]
```

```python
# Set the data and initialization of parameters for the DLM

# Define the data
data = {
                    'd':d,
                    's':s,
                    'N':N,
                    'nreg':nreg,
                    'regressors':monthly_regressors,
                    'seasonal_GP_prior':C_GP_prior,
                    'S':10.,
                    'sigma_trend_prior':0.01,
                    'sigma_seas_prior':0.01,
                    'sigma_AR_prior':0.5
                }

# Initialization
initial_state = {
                 'sigma_trend':0.0001,
                 'sigma_seas':0.001,
                 'sigma_AR':0.01,
                 'rhoAR':0.1,
                }

# Find the maximum likelihood estimator
mle = model_seasonal_ar1.optimizing(data=data, init=[initial_state])

# Update the initial state for the MCMC chain to the maximum likelihood values just obtained
initial_state['sigma_trend'] = mle['sigma_trend']
initial_state['sigma_seas'] = mle['sigma_seas']
initial_state['sigma_AR'] = mle['sigma_AR']
initial_state['rhoAR'] = mle['rhoAR']
```

```python
# Run the HMC DLM (and time it)
start = time.time()
fit = model_seasonal_ar1.sampling(data=data, iter=2000, warmup=1000, chains=1, init = [initial_state], verbose=True, pars=('sigma_trend', 'sigma_seas', 'sigma_AR', 'rhoAR', 'trend', 'beta'))
print(time.time() - start)
```

```python
# Plot recovered trend against the data
plt.plot(T, d, lw = 2, alpha = 0.2)
plt.plot(T, np.mean(fit.extract()['trend'][:,:], axis = 0), color = 'red', ls = '--')
plt.fill_between(T, np.mean(fit.extract()['trend'][:,:], axis = 0) - np.std(fit.extract()['trend'][:,:], axis = 0), np.mean(fit.extract()['trend'][:,:], axis = 0) + np.std(fit.extract()['trend'][:,:], axis = 0), color = 'grey', alpha = 0.5)
plt.show()
```
