```python
# Import stuff
import pystan
import matplotlib.pyplot as plt
import numpy as np
import pickle
import time
import scipy.optimize as optimize
import scipy.interpolate as interpolate
import netCDF4
import mlr
%matplotlib inline
```

```python
code = """

data {
    int N; // number of time steps
    int nreg; // number of regressors
    vector[N] d; // n-composites vectors of N-time-steps -- the data
    vector[N] s; // corresponding standard deviation for each data point
    vector[nreg] regressors[N]; // nreg regressors of N-time-steps
    matrix[nreg, nreg] C_GP_prior; // prior covariance on regressor coefficients
    matrix[4,4] G; // Seasonal transition matrix
    real<lower=0> sigma_trend_prior;
    real<lower=0> sigma_seas_prior;
    real<lower=0> sigma_AR_prior;
}

parameters {
    vector[nreg] b; // regression coefficients
    vector[N] ar; // ar-process
    vector[4] seasonal_components[N]; // seasonal cycle
    vector[N] slope; // dynamical linear trend [slope]
    real x0;
    real<lower=0> sigma_trend;
    real<lower=0> sigma_seas;
    real<lower=0> sigma_AR;
    real<lower=0, upper=1> rho;
}

transformed parameters{
    vector[N] trend; // dynamical linear trend
    vector[N] seasonal; // seasonal cycle
    vector[N] regression_model; // complete regression model

    // Model for the latent dynamical model
    trend[1] = x0;
    seasonal[1] = seasonal_components[1][1] + seasonal_components[1][3];
    regression_model[1] = dot_product(b, regressors[1]);
    for(t in 2:N){

        trend[t] = trend[t-1] + slope[t];
        seasonal[t] = seasonal_components[t][1] + seasonal_components[t][3];
        regression_model[t] = dot_product(b, regressors[t]);
    }
}

model {
 
    // Priors on hyper parameters
    sigma_trend ~ normal(0, sigma_trend_prior);
    sigma_seas ~ normal(0, sigma_seas_prior);
    sigma_AR ~ normal(0, sigma_AR_prior);
    seasonal_components[1] ~ normal(0, 10);
    rho ~ uniform(0, 1);
    b ~ multi_normal(rep_vector(0, nreg), C_GP_prior);
    
    // Prior on AR process initial value
    ar[1] ~ normal(0, sqrt(sigma_AR^2/(1-rho^2)));

    // Model for the latent dynamical model
    for(t in 2:N){
    
        // AR-process
        ar[t] ~ normal(ar[t-1]*rho, sigma_AR);
        
        // Seasonal cycle
        seasonal_components[t] ~ normal(G*seasonal_components[t-1], sigma_seas);

        // Stochastic DLM trend
        slope[t] ~ normal(slope[t-1], sigma_trend);
    }

    # Construct target density
    for(t in 1:N){
        target += normal_lpdf(ar[t] + trend[t] + seasonal[t] + regression_model[t] | d[t], s[t]);
    }
}

generated quantities{

    vector[N] residuals;
    residuals = d - regression_model - seasonal - trend;
}

"""
```

```python
# Compile the model
model = pystan.StanModel(model_code=code)
```

```python
# Import data
data = netCDF4.Dataset('data/BASIC_V1_2016_lotus_seascyc.nc')

# Which pressure and latitude index do you want to analyze?
pressure = 9
latitude = 11

# Time, pressure and latitude variables
T = data['time'][:]
P = data['pressure'][:]
L = data['latitude'][:]

# How many time steps are there?
N = len(T)

# Pick out a pressure and latitude panel: this is the "data" time-series
d = data['o3'][:, pressure, latitude]

# Extract the error on the time-series
s = data['o3_sigma'][:, pressure, latitude]

# Re-scale the data so it is -0.5 to 0.5, apply same scaling transform to error bars
scale = max(d) - min(d)
d = d - np.mean(d)
d = d/scale
s = s/scale

# Plot the data and error bars
plt.title('Pressure {} hPa; Latitude {} deg'.format(P[pressure], L[latitude]))
plt.plot(T, d, color = 'red')
plt.fill_between(T, d - s, d + s, color = 'grey', alpha = 0.5)
plt.xlabel('time, T [Julian date]')
plt.ylabel('Ozone')
plt.show()
```

```python
# Import and rescale all the regressors, extract the bit over the time period you actually want

# ENSO
y = np.loadtxt('regressors/ENSO_MEI_ENSO_1950_201704.dat')
t = np.loadtxt('regressors/ENSOTime_MEI_ENSO_1950_201704.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
enso = Y(T) # This extacts the regressor at the times you have in your dataset, interpolated in case the time grid is different

# Solar
y = np.loadtxt('regressors/Flux_F30_monthly_195111_201706_absolute.dat')
t = np.loadtxt('regressors/Time_F30_monthly_195111_201706_absolute.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
solar = Y(T)

# QBO30
y = np.loadtxt('regressors/multi_qbo30_1953_2016.dat')
t = np.loadtxt('regressors/multi_qbotime_1953_2016.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
qbo30 = Y(T) 

# QBO50
y = np.loadtxt('regressors/multi_qbo50_1953_2016.dat')
t = np.loadtxt('regressors/multi_qbotime_1953_2016.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
qbo50 = Y(T) # This extacts the regressor at the times you have in your dataset, interpolated in case the time grid is different

# SAOD
y = np.loadtxt('regressors/Reg_SAODak_1511979_15122015.dat')
t = np.loadtxt('regressors/Time_SAODak_1511979_15122015.dat')
y = y - np.mean(y)
y = y/(max(y) - min(y))
Y = interpolate.InterpolatedUnivariateSpline(t, y)
saod = Y(T)

# Stack the regressors into big 2d array
nreg = 5 # number of regressors
regressors = np.column_stack([enso, solar, qbo30, qbo50, saod])

# Plot all the regressors
plt.plot(T, enso)
plt.show()

plt.plot(T, solar)
plt.show()

plt.plot(T, qbo30)
plt.show()

plt.plot(T, qbo50)
plt.show()

plt.plot(T, saod)
plt.show()
```

```python
# Declare matrix for seasonal cycle updates
G = np.zeros((4,4))
G[0:2, 0:2] = np.array([[np.cos(2*np.pi/12), np.sin(2*np.pi/12)],[-np.sin(2*np.pi/12), np.cos(2*np.pi/12)]])
G[2:4, 2:4] = np.array([[np.cos(2*2*np.pi/12), np.sin(2*2*np.pi/12)],[-np.sin(2*2*np.pi/12), np.cos(2*2*np.pi/12)]])
```

```python
# Create seasonal GP
x = np.arange(1,13)
l = 1.4
A = 1

# Covariance
C = np.zeros((12, 12))
for i in range(0, 12):
    for j in range(0,12):
        C[i,j] = A*np.exp(-(min(abs(i-j), abs(i-j-12), abs(i-j+12)))**2/(2*l**2))
        
L = np.linalg.cholesky(C)
plt.imshow(C)
plt.show()

for i in range(0, 3):
    Y = interpolate.InterpolatedUnivariateSpline(x, np.dot(L, np.random.normal(0, 1, 12)), k = 3)
    plt.plot(np.linspace(1, 12, 100), Y(np.linspace(1, 12, 100)))
plt.show()

C_GP_prior = np.kron(np.eye(nreg), C)
plt.imshow(C_GP_prior)
plt.show()
```

```python
# Construct seasonal regressors

# Compute the mask
mask = np.zeros((12, len(T)))
for i in range(0, 12):
    mask[i,np.arange(0,len(T)) % 12 == i] = 1

# Set-up the masked regressors
monthly_regressors = np.zeros((len(T), nreg*12))
for r in range(nreg):
    for i in range(0, 12):
        monthly_regressors[:, r*12 + i] = regressors[:,r]*mask[i,:]
```

```python
# Do a cheeky MLR for initializing the DLM
ar_process, beta_coefficients, trend_slope, seasonal_cycle, sigma_AR, rho_AR = mlr.mlr_initialization(regressors.T, nreg, T, d)
beta_coefficients = np.kron(beta_coefficients[0:nreg], np.ones(12))
```

```python
# Set the data and initialization of parameters for the DLM

# Define the data
data = {
                    'd':d,
                    's':s,
                    'N':N,
                    'nreg':nreg*12,
                    'regressors':monthly_regressors,
                    'G':G,
                    'C_GP_prior':C_GP_prior,
                    'sigma_trend_prior':1e-4,
                    'sigma_seas_prior':0.01,
                    'sigma_AR_prior':0.5
                }

# Initialization
initial_state = {'ar':ar_process,
                 'b':beta_coefficients, 
                 'slope':trend_slope, 
                 'x0':0,
                 'seasonal_components':seasonal_cycle,
                 'sigma_trend':0.0001,
                 'sigma_seas':0.001,
                 'sigma_AR':sigma_AR,
                 'rho':rho_AR
                }

# Find the maximum likelihood estimator
mle = model.optimizing(data=data, init=[initial_state])

# Update the initial state for the MCMC chain to the maximum likelihood values just obtained
initial_state['sigma_trend'] = mle['sigma_trend']
initial_state['sigma_seas'] = mle['sigma_seas']
initial_state['sigma_AR'] = mle['sigma_AR']
initial_state['rho'] = mle['rho']
initial_state['seasonal_components'] = mle['seasonal_components']
initial_state['x0'] = mle['x0']
initial_state['slope'] = mle['slope']
initial_state['b'] = mle['b']
initial_state['ar'] = mle['ar']
```

```python
# Do the sampling
start = time.time()
fit = model.sampling(data=data, iter=4000, warmup=2000, chains=1, init = [initial_state])
print(time.time() - start)
```

```python
# Plot recovered trend
plt.plot(T, np.mean(fit.extract()['trend'][:,:], axis = 0), color = 'red', ls = '--')
plt.fill_between(T, np.mean(fit.extract()['trend'][:,:], axis = 0) - np.std(fit.extract()['trend'][:,:], axis = 0), np.mean(fit.extract()['trend'][:,:], axis = 0) + np.std(fit.extract()['trend'][:,:], axis = 0), color = 'grey', alpha = 0.5)
plt.show()
```
