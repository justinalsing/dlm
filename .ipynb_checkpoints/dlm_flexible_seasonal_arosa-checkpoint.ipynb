```python
# Import stuff
import pystan
import matplotlib.pyplot as plt
import numpy as np
import pickle
import time
import scipy.optimize as optimize
import scipy.interpolate as interpolate
import netCDF4
from models.stan_dlm_models import *
%matplotlib inline
```

```python
# Compile the seasonal DLM model
model_seasonal = pystan.StanModel(model_code=code_seasonal)

# Compile the DLM model
model_kalman_ar1 = pystan.StanModel(model_code=code_kalman_ar1)
```

```python
# Import data

# AROSA Data
yaro = np.loadtxt('data/aro_197901_201712.txt')
taro = np.loadtxt('data/aro_time_197901_201712.txt')

# Extract time, pressure and latitude variables from the netCDF
T = taro#[0:-36]
O = yaro#[0:-36]

P = [1]
L = [45] #data['latitude'][:]

# How many time steps are there? (ie how long is the time-series)
N = len(T)

print, P, L

# Which pressure and latitude index do you want to analyze?
pressure = 0
latitude = 0
pind = 0
lind = 0

# Pick out a pressure and latitude panel: this is the "data" time-series
d = O
s = O*0.+1e-20

# How many time steps are there?
N = len(T)

# Re-scale the data so it is -0.5 to 0.5, apply same scaling transform to error bars
scale = max(d) - min(d)
d = d - np.mean(d)
d = d/scale
s = s/scale

# Plot the data and error bars
plt.title('Pressure {} hPa; Latitude {} deg'.format(P[pressure], L[latitude]))
plt.plot(T, d, color = 'red')
plt.fill_between(T, d - s, d + s, color = 'grey', alpha = 0.5)
plt.xlabel('time, T [Julian date]')
plt.ylabel('Ozone')
plt.show()
```

```python
# Import the regressors

# Again you can do this however you want, but the result must be the following variables loaded into python
# nreg (int) = number of regressors
# regressors float[N, nreg] = 2d array with each column representing one regressor over the times corresponding to
# the data to be analysed

# IMPORTANT: Missing values/NaNs are NOT supported, please interpolate missing values or similar so they are all 
# real valued

# ENSO
regressor_data = np.loadtxt('regressors/enso_mei_1950_201802.txt')
Y = interpolate.InterpolatedUnivariateSpline(regressor_data[:,0], regressor_data[:,1])
enso = Y(T)

# SOLAR
regressor_data = np.loadtxt('regressors/flux_F30_monthly_195111_201803_absolute.txt')
Y = interpolate.InterpolatedUnivariateSpline(regressor_data[:,0], regressor_data[:,1])
solar = Y(T)

# QBO30
regressor_data = np.loadtxt('regressors/multi_qbo30_1953_2018.txt')
Y = interpolate.InterpolatedUnivariateSpline(regressor_data[:,0], regressor_data[:,1])
qbo30 = Y(T)

# QBO50
regressor_data = np.loadtxt('regressors/multi_qbo50_1953_2018.txt')
Y = interpolate.InterpolatedUnivariateSpline(regressor_data[:,0], regressor_data[:,1])
qbo50 = Y(T)

# SAOD
regressor_data = np.loadtxt('regressors/sad_1979_2017_10deg_60S_60N.txt')
Y = interpolate.InterpolatedUnivariateSpline(regressor_data[:,0], regressor_data[:,10])
saod = Y(T)

# AO
regressor_data = np.loadtxt('regressors/ao_197901_201802.txt')
Y = interpolate.InterpolatedUnivariateSpline(regressor_data[:,0], regressor_data[:,1])
ao = Y(T)

# AAO
regressor_data = np.loadtxt('regressors/aao_197901_201802.txt')
Y = interpolate.InterpolatedUnivariateSpline(regressor_data[:,0], regressor_data[:,1])
aao = Y(T)

# Stack the regressors into big 2d array
nreg = 6 # number of regressors
regressors = np.column_stack([enso, solar, qbo30, qbo50, saod, ao]) # Stack of all the regressors together in a 2d array

# Plot all the regressors to check they look OK
plt.plot(T, enso)
plt.title('ENSO')
plt.xlabel('time, T [Julian date]')
plt.ylabel('[re-scaled units]')
plt.show()

plt.plot(T, solar)
plt.title('Solar F30')
plt.xlabel('time, T [Julian date]')
plt.ylabel('[re-scaled units]')
plt.show()

plt.plot(T, qbo30)
plt.title('QBO30')
plt.xlabel('time, T [Julian date]')
plt.ylabel('[re-scaled units]')
plt.show()

plt.plot(T, qbo50)
plt.title('QBO50')
plt.xlabel('time, T [Julian date]')
plt.ylabel('[re-scaled units]')
plt.show()

plt.plot(T, saod)
plt.title('SAOD')
plt.xlabel('time, T [Julian date]')
plt.ylabel('[re-scaled units]')
plt.show()

plt.plot(T, ao)
plt.title('AO')
plt.xlabel('time, T [Julian date]')
plt.ylabel('[re-scaled units]')
plt.show()

plt.plot(T, aao)
plt.title('AAO')
plt.xlabel('time, T [Julian date]')
plt.ylabel('[re-scaled units]')
plt.show()
```

```python
# Declare matrix for seasonal cycle updates
G = np.zeros((4,4))
G[0:2, 0:2] = np.array([[np.cos(2*np.pi/12), np.sin(2*np.pi/12)],[-np.sin(2*np.pi/12), np.cos(2*np.pi/12)]])
G[2:4, 2:4] = np.array([[np.cos(2*2*np.pi/12), np.sin(2*2*np.pi/12)],[-np.sin(2*2*np.pi/12), np.cos(2*2*np.pi/12)]])
```

```python
# Create seasonal Gaussian Process prior 
x = np.arange(1,13)
l = 1.4
A = 1

# Covariance
C = np.zeros((12, 12))
for i in range(0, 12):
    for j in range(0,12):
        C[i,j] = A*np.exp(-(min(abs(i-j), abs(i-j-12), abs(i-j+12)))**2/(2*l**2))

C_GP_prior = C
L_GP_prior = np.linalg.cholesky(C_GP_prior)
#C_GP_prior = np.kron(np.eye(nreg), C)
plt.imshow(C_GP_prior)
plt.show()
```

```python
# Construct seasonal regressors

# Compute the mask
mask = np.zeros((12, len(T)))
for i in range(0, 12):
    mask[i,np.arange(0,len(T)) % 12 == i] = 1

# Set-up the masked regressors
monthly_regressors = np.zeros((len(T), nreg*12))
for r in range(nreg):
    for i in range(0, 12):
        monthly_regressors[:, r*12 + i] = regressors[:,r]*mask[i,:]
```

```python
# Set the data and initialization of parameters for the DLM

# Define the data
data_seasonal = {
                    'd':d,
                    's':s,
                    'N':N,
                    'nreg':nreg,
                    'regressors':regressors,
                    'G':G,
                    'L_GP_prior':L_GP_prior,
                    'sigma_trend_prior':1e-4,
                    'sigma_seas_prior':0.01,
                    'sigma_AR_prior':0.5
                }
```

```python
# Do a normal DLM run to initialize the seasonal DLM

# Set the data and initialization of parameters for the DLM

# Define the data
data_kalman = {
                    'time_series':d,
                    'stddev':s,
                    'N':N,
                    'nreg':nreg,
                    'regressors':regressors,
                    'S':10.,
                    'sigma_trend_prior':1e-4,
                    'sigma_seas_prior':0.01,
                    'sigma_AR_prior':0.5
                }

# Initialization
initial_state = {
                 'sigma_trend':0.0001,
                 'sigma_seas':0.001,
                 'sigma_AR':0.01,
                 'rhoAR':0.1,
                }

# Run the HMC DLM (and time it)
fit_kalman = model_kalman_ar1.sampling(data=data_kalman, iter=1000, warmup=500, chains=1, init = [initial_state], verbose=True, pars=('sigma_trend', 'sigma_seas', 'sigma_AR', 'rhoAR', 'slope', 'seasonal_components', 'beta', 'trend', 'ar'))
```

```python
# Set initialization for seasonal model
initial_state = {'ar':np.mean(fit_kalman.extract()['ar'], axis=0),
                 'b':np.array([np.mean(fit_kalman.extract()['beta'], axis=0) for i in range(12)]).T, 
                 'slope':np.mean(fit_kalman.extract()['slope'], axis=0), 
                 'x0':np.mean(fit_kalman.extract()['trend'][:,0]),
                 'seasonal_components':np.mean(fit_kalman.extract()['seasonal_components'], axis=0),
                 'sigma_trend':np.mean(fit_kalman.extract()['sigma_trend']),
                 'sigma_seas':np.mean(fit_kalman.extract()['sigma_seas']),
                 'sigma_AR':np.mean(fit_kalman.extract()['sigma_AR']),
                 'rho':np.mean(fit_kalman.extract()['rhoAR'])
                }
```

```python
# Optimize from the above starting point
opt = model_seasonal.optimizing(data=data_seasonal, init = [initial_state])#, algorithm='Newton', iter=50)
```

```python
# Do the sampling
start = time.time()
fit_seasonal = model_seasonal.sampling(data=data_seasonal, iter=10000, warmup=2000, chains=1, init = [initial_state])
print(time.time() - start)
```

```python
with open(samples_file) as f:
    found_header = False
    header_length = 0
    for i, line in enumerate(f):
        line = line.strip()
        if found_header == False and line[0] != '#':
            labels = line.split(',')
            found_header = True
        elif found_header == True and line[0] != '#':
            header_length = i
            break

samples = np.loadtxt(samples_file, skiprows = header_length, delimiter = ',')

from collections import OrderedDict, Counter
from collections import Callable, Iterable

counter = Counter()
        
for i, label in enumerate(labels):
    tokens = label.split('.')
         
    if len(tokens) == 1:
        counter[label] += 1
    else:
        tlabel = "".join(label.split('.')[:-1])
        counter[tlabel] += 1
            
output = {}
       
for tlabel, count in counter.items():
    idxs = np.where([label.startswith(tlabel) for label in labels])
    if count == 1:
        output[tlabel] = samples[:, idxs].reshape(-1)
    else:
        output[tlabel] = samples[:, idxs].reshape(-1, count)
```
